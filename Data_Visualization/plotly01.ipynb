```python
df = px.data.gapminder()
df.head()
```

```python
# Creating a bubble chart with query exact year
px.scatter(df.query('year == 2007'), x = 'gdpPercap', y = 'lifeExp', size = 'pop',
           color = 'continent', log_x = True, hover_name = 'country',
           size_max = 60)
```

```python
# Creating a bubble chart without query but with animation for years
px.scatter(df, x = 'gdpPercap', y = 'lifeExp', size = 'pop',
           color = 'continent', log_x = True, hover_name = 'country',
           size_max = 60, animation_frame = 'year', range_y=[25,90],
           facet_col='continent')
```

```python
# Making the line chart for one country
px.line(df.query('country == "Poland"'), x = 'year', y = 'pop')
```

```python
# Making the line chart for one continent
px.line(df.query('continent == "Europe"'), x = 'year', y = 'pop',
        color = 'country')
```

```python
# Using the GitHub GIST
def fetch_financial_data(company = 'AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df = fetch_financial_data()
df = df.reset_index()
df.head()
```

```python
# Making the Area Chart
df = df[df['Date'] >= '2019-01-01']
px.area(df, x = 'Date', y = 'Close', log_y = True)
```

```python
from seaborn import load_dataset
import plotly.express as px

df = load_dataset('diamonds')
df.head()
```

```python
# Changing the data of column carat into the provided list
# 10 values (abc...)
df.carat = pd.cut(df.carat, bins = 10, labels = list('abcedfghij'))
df.head()
```

```python
df = pd.read_csv("https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/plotly-course/us-cities-top-1k.csv")
df.head()
```

```python
# Scatter map box chart (defuning function to call)
def plot_map(mapbox_style):
  fig = px.scatter_mapbox(df, lat = 'lat', lon = 'lon', hover_name = 'City',
                          hover_data = ['State', 'Population'], zoom = 3)
  fig.update_layout(mapbox_style = mapbox_style,
                    margin = {'r': 10, 't': 10, 'l': 10, 'b': 10})
  fig.show()
```

```python
plot_map('carto-positron')
```
