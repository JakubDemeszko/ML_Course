```python
def fetch_financial_data(company = 'AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df_amzn = fetch_financial_data(company = 'AMZN')
df_apl = fetch_financial_data(company = 'AAPL')
df_msft = fetch_financial_data(company = 'MSFT')
df_nflx = fetch_financial_data(company = 'NFLX')
```

```python
import pandas as pd

# Connecting the current DFs

df = pd.concat([df_amzn.Close, df_apl.Close, df_msft.Close, df_nflx.Close], axis = 1, join = 'inner')
df.columns = ['Amazon', 'Apple', 'Microsoft', 'Netflix']
df = df.reset_index()
df.head()
```

```python
# Creating a chart for all companies
import plotly.graph_objects as go

trace0 = go.Scatter(x = df.Date,
                    y = df.Amazon,
                    name = 'Amazon')
trace1 = go.Scatter(x = df.Date,
                    y = df.Apple,
                    name = 'Apple')
trace2 = go.Scatter(x = df.Date,
                    y = df.Microsoft,
                    name = 'Microsoft')
trace3 = go.Scatter(x = df.Date,
                    y = df.Netflix,
                    name = 'Netflix')
data = [trace0, trace1, trace2, trace3]
fig = go.Figure(data = data)
fig.show()
```

```python
# Creating a loop to get the same chart
data = [go.Scatter(x = df.Date, y = df[col], name = col) for col in df.columns[1:]]
fig = go.Figure(data = data)
fig.show()
```
