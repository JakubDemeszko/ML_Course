```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
```

```python
# Training data are the data which are used to train the model and then after training we can test it on the test data
# Other way: Data divided by: a) X_Train, b) X_Val (validation), c) X_test

np.random.seed(42)
np.set_printoptions(edgeitems = 30, linewidth=100000,
                    formatter = dict(float=lambda x: f'{x: 2f}'))
sns.set()
```

```python
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(f'{data[:5]}\n')
print(target[:5])
```

```python
all_data = np.c_[data, target]
all_data[:5]
```

```python
df = pd.DataFrame(data = all_data, columns = raw_data.feature_names + ['target'])
df.head()
```

```python
df.describe().T.apply(lambda x: round(x, 2))
```

```python
df.target.value_counts()
```

```python
# Separating from the DF the last column target.
data = df.copy()
target = data.pop('target')
data.head()
```

```python
target.head()
```

```python
# Splitting the data to the Train and Test samples
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data, target, random_state = 42)
```

```python
# Setting up the target value for the same values
X_train, X_test, Y_train, Y_test = train_test_split(data, target, random_state = 42, test_size = 0.1, stratify = target)
Y_train.value_counts()
```
