```python
import pandas as pd
import numpy as np

# Creating a basic DF.
df = pd.DataFrame(data = [[12,12,12], [2,3,5]], index = ['first', 'second'],
                  columns = ['Col_1', 'Col_2', 'Col_3'])
df
```

```python
# Getting data from choosen column
df[df['Col_1'] > 3]
```

```python
# Alternative way
df.Col_1
```

```python
# Using the LOC with term
df.loc['first'] > 13
```

```python
def fetch_financial_data(company = 'AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df = fetch_financial_data()
df
```

```python
# Calculating the new column
df['Average'] = (df.Open + df.Close) / 2
df.head()
```

```python
# Filtering the data.
df.Close > 169
```

```python
# Another way
df[df.Close > 169]
```

```python
# Filtering in Index
df[(df.index > '2020-01-01') & (df.index < '2024-02-05')]
```
