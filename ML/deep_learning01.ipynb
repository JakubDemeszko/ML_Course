```python
# Siec neuronowa sklada sie z: wagi, elementy modelu, ktore trenujemy.
# Funkcja aktywacji, funkcja straty (kontrola bledu), optymalizator (pozwala uczyc sie sieci)
# Przewidywanie liczb na podstawie modelu

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from tensorflow.keras.datasets.mnist import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

np.set_printoptions(precision = 12, suppress = True, linewidth = 120)
print(tf.__version__)
```

```python
# Przygotowanie i zaladowanie danych
(X_train, y_train), (X_test, y_test) = load_data()
print(X_train[0])
```

```python
X_train = X_train / 255.
X_test = X_test / 255.
```

```python
# Budowa sieci neuronowej. Warstwy dodajemy metoda Add. Flatten - wyplaszcza nasze dane (by siec wiedziala czego sie spodziwac)
# Warstwa Dens - zawiera 128 neuronow
# Warstwa Dropout - porzuci 20% neuronow w warstwie
# Dodanie ponownie dens - 10 neuronow (tyle ile klas), aktywacja softmax - prawdopodobienstwo poszczegolnej klasy
# Kompilacja modelu
model = Sequential()
model.add(Flatten(input_shape = (28,28)))
model.add(Dense(units=128, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units=10, activation = 'softmax'))

model.compile(optimizer='Adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
# Trenowanie modelu
history = model.fit(X_train, y_train, epochs = 5)
```

```python
# Ocena modelu
model.evaluate(X_test, y_test, verbose = 2)
```

```python
metrics = pd.DataFrame(history.history)
metrics
```

```python
# Wizualizacja - chcemy by strata nam malala, a accuracy roslo z kazda kolejna epoka
fig = make_subplots(rows = 2, cols = 1)
fig.add_trace(go.Scatter(y = metrics['loss'], name = 'loss'), row = 1, col = 1)
fig.add_trace(go.Scatter(y = metrics['accuracy'], name = 'accuracy'), row = 2, col = 1)
fig.update_layout(width = 800, height = 400)
```

```python
# Predykcja na podstawie modelu
model.predict(X_test)
```

```python
# Zwrocenie konkretnej klasy
y_pred_probabilities = model.predict(X_test)
y_pred = y_pred_probabilities.argmax(axis=-1)

print(y_pred)
```

```python
# Laczenie danych
pred = pd.concat([pd.DataFrame(y_test, columns=['y_test']), pd.DataFrame(y_pred, columns=['y_pred'])], axis = 1)
pred.head()
```

```python
# Wyciagniecie tego, gdzie jest zle
missclassified = pred[pred['y_test'] != pred['y_pred']]
missclassified.head()
```
